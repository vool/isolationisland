Data downloaded from [National Monuments Service - underwater-archaeology/wreck-viewer](https://dahg.maps.arcgis.com/apps/webappviewer/index.html?id=89e50518e5f4437abfa6284ff39fd640)


```python
wrecks = pd.read_csv('data/Wrecks_April_2018.csv',encoding='ISO-8859-1')
```

```python
wrecks.tail()
```

```python
# remove unneeded columns
smr_cleaned = smr.drop(['ENTITY_ID','NAT_GRID_E','NAT_GRID_N','ITM_E', 'ITM_N'], axis = 1)

# rename coords
smr_cleaned.rename(columns = {'Lat':'lat', 'Lon':'lng'}, inplace = True)

#set name
smr_cleaned['title'] = smr_cleaned['CLASSDESC']

#set type
smr_cleaned['type'] = 'smr'

smr_cleaned.head()
```

```python
niah = pd.read_csv('data/NIAH161116.csv',encoding='ISO-8859-1')
```

```python
niah.head()
```

```python
# remove unneeded columns
niah_cleaned = niah.drop(['COUNTY_ID', 'PLANAUTH','X_COORD','Y_COORD','RATING', 'X_COORD_ITM', 'Y_COORD_ITM', 'SURVEY_ID'], axis = 1)

# rename coords
niah_cleaned.rename(columns = {'LATITUDE':'lat', 'LONGITUDE':'lng'}, inplace = True)

#set name
smr_cleaned['title'] = niah_cleaned['NAME']

#set type
niah_cleaned['type'] = 'niah'

niah_cleaned.head()
```

```python
# remove unneeded columns
wrecks_cleaned = wrecks.drop(['OBJECTID','Source_of_Co_ordinate', 'Record_Source', 'Date_of_Loss_Year_Only'], axis = 1)

# rename coords and name 
wrecks_cleaned.rename(columns = {'DD_Lat':'lat', 'DD_Long':'lng', 'Wreck_Name': 'name'}, inplace = True)

#drop wrecks at Null Island
wrecks_cleaned =  wrecks_cleaned[wrecks_cleaned.lat != 0.0]

#set type
wrecks_cleaned['type'] = 'wrecks'

wrecks_cleaned.head()
```
